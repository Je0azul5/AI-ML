Importaciones de bibliotecas

```
import pandas as pd
import sweetviz as sv
```

Se implementa el dataframe en un repositorio en GitHub german_credit.csv 

```
dfm=pd.read_csv("https://raw.githubusercontent.com/Je0azul5/weired/master/german_credit_data.csv")
dfm
```

Se revisan si hay campos vacios

```
pd.isna(dfm)
```

Se describe analiticamente caracteristicas del DataFrame

```
dfm.describe()
```

Se ordera de forma asendente por edad, en donde se revisa si solo la edad influye en el prestamo de dinero

```
dfm.sort_values(by="Age")
```

Se revisa, edad, cantidad prestada, y el tipo de fortuna del cliente, ordenados por edad

```
dfm[['Age','Credit amount','Saving accounts']].sort_values(by="Age")
```

Se revisa, edad, cantidad prestada, y si tiene vivienda propia del cliente, ordenados por edad

```
dfm[['Age','Credit amount','Housing']].sort_values(by="Age")
```

Se revisa, el riesgo de los clientes por sexo

```
dfm.groupby('Risk').get_group('good').sort_values(by="Sex")
```

Se ordena por sexo femenino y el proposito tiene cada cliente

```
dfm.groupby('Sex').get_group('female').sort_values(by="Purpose")
```

Se ordena por sexo femenino y la cantidad de credito aprovada

```
dfm.groupby('Sex').get_group('female').sort_values(by="Credit amount")
```

Se ordena por sexo masculino y el proposito tiene cada cliente

```
dfm.groupby('Sex').get_group('male').sort_values(by="Purpose")
```

Se ordena por sexo masculino y la cantidad de credito aprovada

```
dfm.groupby('Sex').get_group('male').sort_values(by="Credit amount")
```

Se utiliza SweerViz para generar un reporte mas detallado

```
my_report = sv.analyze(dfm)
my_report.show_html()
```
