```
import random

#Copia del tablero actual
def copiaTablero(board):
	duplicado = []

	for i in board:
		duplicado.append(i)

	return duplicado

#Imprime el tablero del juego 
def dibujar(board):
  ghj=[]
  r=0
  for i in board:
    if i =='':
      
      ghj.append(' ')
    else:
      ghj.append(i)
    r+=1
  print('' + ghj[1] + '│' + ghj[2] + '│' + ghj[3])
  print('------')
  print(''+ ghj[4] + '│' + ghj[5] + '│' + ghj[6])
  print('------')
  print(''+ ghj[7] + '│' + ghj[8] + '│' + ghj[9])

#Selecciona la ficha con la cual se va a jugar 
def seleccionFicha():
	ficha = ''
	while not(ficha == 'X' or ficha == 'O'):
		print('¿Quiere ser X - O?')
		ficha = input().upper()
		if (ficha != 'X' and ficha != 'O'):
			print("Debe seleccionar X o O")
	if ficha == 'X':
		return ['X','O']
	else:
		return ['O','X']

#Escoje el jugador inicial de forma aleatoria
def seleccionInicial():
	if random.randint(0, 1) == 0:
		return 'computador'
	else:
		return 'jugador'
#Mueve la ficha en funcion del movimiento realizado
def moverFicha(board, ficha, move):
	board[move] = ficha

#Verifica si es tablero es un estado ganador
def esGanador(brd, let):

		if (brd[7] == let and brd[8] == let and brd[9] == let):
			return True;
		if (brd[4] == let and brd[5] == let and brd[6] == let):
			return True;
		if (brd[1] == let and brd[2] == let and brd[3] == let):
			return True;
		if (brd[7] == let and brd[4] == let and brd[1] == let):
			return True;
		if (brd[8] == let and brd[5] == let and brd[2] == let):
			return True;
		if (brd[9] == let and brd[6] == let and brd[3] == let):
			return True;
		if (brd[7] == let and brd[5] == let and brd[3] == let):
			return True;
		if (brd[9] == let and brd[5] == let and brd[1] == let):
			return True;
		return False;

#Retorna True si el espacio a jugar es posible y en caso contrario False
def espacioLibre(board, move):
	if (board[move] == ''):
		return True
	else:
		return False

#Retorna el numero de la casilla seleccionada si esta libre 
def movimientoJugador(board):
	move = ''
	while move not in '1 2 3 4 5 6 7 8 9'.split() or not espacioLibre(board, int(move)):
		print('¿Cual va a ser su siguiente moviemiento? (1-9)')
		move = input();
		if (move not in '1 2 3 4 5 6 7 8 9'):
			print ("Movimento invalido, elija un numero del 1 al 9")
	return int(move)

#Retorna True si todos los espacios han sido ocupados
def tableroLleno(board):
	for i in range(1, 10):
		if espacioLibre(board, i):
			return False
	return True

#Retorna una lista con los valores posibles 
def opcionesPosibles(board):
	opciones = []
	for i in range(1, 10):
		if espacioLibre(board, i):
			opciones.append(i)
	return opciones

#Verifica si hay algun ganador, en caso contrario si el tablero esta lleno
def esFinal(board, pcLetra):
	#-1 si el jugador gana
	#1 si el pc gana
	#0 empate
	if pcLetra == 'X':
		pjLetra = 'O'
	else:
		pjLetra = 'X'
	if (esGanador(board, pcLetra)):
		return 1
	elif (esGanador(board, pjLetra)):
		return -1
	elif (tableroLleno(board)):
		return 0
	else:
		return None

#Algoritmo alfa beta poda
def alfaBetaPoda(board, pcLetra, turn, alpha, beta):
	if pcLetra == 'X':
		pjLetra = 'O'
	else:
		pjLetra = 'X'

	if turn == pcLetra:
		sigTurn = pjLetra
	else:
		sigTurn = pcLetra

	finish = esFinal(board, pcLetra)

	if (finish != None):
		return finish

	pos = opcionesPosibles(board)

	if turn == pcLetra:
		for move in pos:
			moverFicha(board, turn, move)
			val = alfaBetaPoda(board, pcLetra, sigTurn, alpha, beta)
			moverFicha(board, '', move)
			if val > alpha:
				alpha = val

			if alpha >= beta:
				return alpha
		return alpha

	else:
		for move in pos:
			moverFicha(board, turn, move)
			val = alfaBetaPoda(board, pcLetra, sigTurn, alpha, beta)
			moverFicha(board, '', move)
			if val < beta:
				beta = val

			if alpha >= beta:
				return beta
		return beta

#Elegimos el movimineto de la maquina
def movimientoMaquina(board, turn, pcLetra):
	a = -2
	opciones = []

	if pcLetra == 'X':
		pjLetra = 'O'
	else:
		pjLetra = 'X'


	opcionesPosiblesOn = opcionesPosibles(board)

	for move in opcionesPosiblesOn:
		moverFicha(board, pcLetra, move)
		val = alfaBetaPoda(board, pcLetra, pjLetra, -2, 2)		
		moverFicha(board, '', move)
		if val > a:
			a = val
			opciones = [move]

		elif val == a:
			opciones.append(move)

	return random.choice(opciones)
```

```
print('Inicia el juego')

jugar = True


theBoard = [''] * 10
pjLetra, pcLetra = seleccionFicha()
turn = seleccionInicial()
print('O ' + turn +' juega primero,')
enJuego = True

while enJuego:
	if turn == 'jugador':
		dibujar(theBoard)
		move = movimientoJugador(theBoard)
		moverFicha(theBoard, pjLetra, move)

		if esGanador(theBoard, pjLetra):
			dibujar(theBoard)
			print('Has ganado')
			enJuego = False
			
		else:
			if tableroLleno(theBoard):
				dibujar(theBoard)
				print('Empate')
				break
			else:
				turn = 'computador'

	else:
		move = movimientoMaquina(theBoard, pjLetra, pcLetra)
		moverFicha(theBoard, pcLetra, move)

		if esGanador(theBoard, pcLetra):
			dibujar(theBoard)
			print("Computador gano")
			enJuego = False

		else:
			if tableroLleno(theBoard):
				dibujar(theBoard)
				print('Empate')
				break
			else:
				turn = 'jugador'

fichaNew = ''
```
